# Transcribing Function

```python
import whisper
import os

# Load Whisper model
model_whisper = whisper.load_model("tiny.en") # Can be other models

# Function to transcribe a single audio file
def transcribe_audio(file_path):
    result = model_whisper.transcribe(file_path)
    return result["text"]
```

# Text Embeddings Function

```python
from openai import OpenAI
import time
from google.colab import userdata

my_sk = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=my_sk) # your secret key

def generate_embeddings(text):
    """
    Generates embeddings for the given text using OpenAI's API.
    """
    response = client.embeddings.create(
      input=text,
      model="text-embedding-3-large" # Choose the model suitable for your needs
    )
    return response.data[0].embedding # This might need adjustment based on the actual API response
```

# Loading Catboost Model

```python
from catboost import CatBoostClassifier
model_cb = CatBoostClassifier()
model_cb.load_model('/content/drive/MyDrive/ADReSSo21/diagnosis/model_cb.dump');
```

# Testing Catboost

```python
test_audio_path = '/content/00360.mp3'

# Assuming functions for transcription and embedding are defined as before
transcription = transcribe_audio(test_audio_path)
embedding = generate_embeddings(transcription)

# Ensure embedding is in the correct shape for prediction
embedding = [embedding]  # Make it a list of one embedding

# Use the loaded model to make a prediction
prediction = model_cb.predict(embedding)

# Output the prediction result
print(f"Prediction: {'Dementia' if prediction[0] == 1 else 'No Dementia'}")
```
